```python
import pandas as pd
download_url=("https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv")
df = pd.read_csv(download_url)
type(df)
```

```python
pd.set_option("display.max.columns",None)
df.head()
```

```python
import matplotlib.pyplot as plt
df.plot(x="Rank", y=["P25th", "Median", "P75th"])
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(df["Rank"],df["P75th"])
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(df["Rank"],df["Unemployment_rate"])
plt.show()
```

```python
median_column=df["Median"]
median_column.plot(kind="hist")
plt.show()
```

```python
median_column=df["College_jobs"]
median_column.plot(kind="hist")
plt.show()
```

```python
top_5 = df.sort_values(by="Median",ascending=False).head()
top_5.plot(x="Major", y="Median", kind="bar", rot=5, fontsize=4)
plt.show()
```

```python
top_medians=df[df["Median"]>70000].sort_values("Median")
top_medians.plot(x="Major", y=["P25th", "Median", "P75th"],kind="bar")
plt.show()
```

```python
df.plot(x="Median", y="Unemployment_rate", kind="scatter")
plt.show()
```

```python
df.plot(x="ShareWomen", y="Median", kind="scatter")
plt.show()
```

```python
df.plot(x="ShareWomen", y="Unemployment_rate", kind="scatter")
plt.show()
```
