```python
from pyspark.sql import SparkSession  
from pyspark.sql.types import StructType, StructField
from pyspark.sql.types import DoubleType, IntegerType, StringType, FloatType


def get_field(t: str):
    if t == "float":
        return FloatType()
    elif t == "integer":
        return IntegerType()
    else:
        raise Exception("Not expected: %s" % t)

struct_fields = []
for index, row in af.iterrows():
    struct_fields.append(StructField(row['name'], get_field(row['type'])))

schema = StructType(struct_fields)
```

```python
scSpark = SparkSession \
    .builder \
    .appName("Python Spark SQL basic example: Reading CSV file without mentioning schema") \
    .getOrCreate()
 
sdfData = scSpark.read.csv("../data/smartphone-activity/dataset.csv", header=True, sep=",", schema=schema)
sdfData.show()
```
