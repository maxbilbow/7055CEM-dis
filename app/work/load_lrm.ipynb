```python
import os
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler

ROOT = r".."
DATA_IN = os.path.join(ROOT, "data", "data-in") #Config.get("path.data.in"))
DATA_FOLDER = "smartphone-activity"
SOURCE_DATASET = os.path.join(DATA_IN ,"dataset.csv")
VECTOR_COL = "features"

def create_df_vector(df):
    columns = list(df.drop("activities").toPandas().columns)
    assembler = VectorAssembler(
        inputCols=columns,
        outputCol=VECTOR_COL)

    df_vector = assembler.transform(df)
    df_vector = df_vector.select(*[VECTOR_COL, "activity"])
    return df_vector


def test_model(model):
    prediction = model.transform(test_data)

    # obtain evaluator.
    evaluator = MulticlassClassificationEvaluator(metricName="accuracy", labelCol="activity")

    # compute the classification error on test data.
    accuracy = evaluator.evaluate(prediction)
    print("Test Error = %g" % accuracy)


spark = SparkSession \
    .builder \
    .appName("Python Spark SQL basic example: Reading CSV file without mentioning schema") \
    .getOrCreate()
dataset = spark.read.load(SOURCE_DATASET, format="csv", sep=",", inferSchema=True, header=True)
dataset.show(1)
```

```python
test_data = create_df_vector(dataset)
print("done")
```
